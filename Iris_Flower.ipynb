```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
np.random.seed(0)
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df.head()
```

```python
df['is_train'] = np.random.uniform(0, 1, len(df)) <= .80
df.head()
```

```python
train, test = df[df['is_train']==True], df[df['is_train']==False]
print("Number of observations in the training data :", len(train))
print("Number of observations in the test data :", len(test))
```

```python
features = df.columns[:4]
features
```

```python
y = pd.factorize(train['species'])[0]
y
```

```python
classif = RandomForestClassifier(n_jobs=2, random_state=0)
classif.fit(train[features], y)
```

```python
classif.predict(test[features])
```

```python
test[features]
```

```python
classif.predict_proba(test[features])[10:20]
```

```python
preds = iris.target_names[classif.predict(test[features])]
preds[0:25]
```

```python
test['species'].head()
```

```python
pd.crosstab (test['species'],preds,rownames=['Actual Species'],colnames=['Predicted Species'] )
```
